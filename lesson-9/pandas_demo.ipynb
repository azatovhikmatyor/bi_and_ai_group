### SQL Server

```cmd
pip install pyodbc
pip install SQLAlchemy
```

````python
from sqlalchemy import create_engine

SERVER_NAME = 'Lenovo'
DATABASE_NAME = 'BikeStores'
DRIVER = 'ODBC+Driver+17+for+SQL+Server'
CONNECTION_STRING = f'mssql+pyodbc://{SERVER_NAME}/{DATABASE_NAME}?trusted_connection=yes&driver={DRIVER}'

ms_sql_engine = create_engine(CONNECTION_STRING)
````

````python
products = pd.read_sql("select * from production.products", con=ms_sql_engine)
products.head()
````

````python
categories = pd.read_sql_table(table_name='categories', schema='production', con=ms_sql_engine)
categories.head() 
````

### Postgre SQL

```cmd
pip install sqlalchemy 
pip install psycopg2
```

````python
from sqlalchemy import create_engine
from getpass import getpass
from urllib.parse import quote_plus

username = "postgres"
password = quote_plus(getpass())
host = "localhost"  # or your database host
port = "5432"       # default PostgreSQL port
database = "store"

postgres_engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")
````

````python
sales = pd.read_sql("SELECT * FROM sales", con=postgres_engine)
sales.head()
````
