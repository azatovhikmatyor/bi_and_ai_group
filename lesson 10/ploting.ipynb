$$
f(X) = 2x^2 + 3x - 1
$$

````python
# x = np.arange(-10, 10, 0.1)
x = np.linspace(-10, 10, 10000)
y = 2*x**2 + 3*x - 1

#RGB
plt.plot(x, y, c='#000000', linestyle='dashed')
plt.show()
````

````python
x = [1, 3, 4, 7, 9]
y = [5, 3, 7, 5, 3]

plt.scatter(x, y, marker='v')
plt.show()
````

$$f(x) = 3*sin(3x + \pi)$$

````python
x = np.linspace(-10, 10, 10000)
f_x = 3*np.sin(3*x + np.pi)

plt.plot(x, f_x)
plt.show()
````
