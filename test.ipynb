```python
import pandas as pd
import requests

url = 'http://universities.hipolabs.com/search?country=United+States'

res = requests.get(url)
res.raise_for_status()
```

```python
data = res.json()
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df[df['domains'].apply(lambda row: len(row)) > 1].head()
```

```python
df_exp = df.explode('domains')
```

```python
df_exp[df_exp['name'].isin(
    df[df['domains'].apply(lambda row: len(row)) > 2]['name']
)]
```

```python
df_exp[df_exp['name'].isin(
    (df_exp.groupby('name')['name'].agg(['count']) > 1).reset_index()['name'].to_list()
)]
```

```python
dff = df_exp.groupby('name', as_index=False)['name'].agg(['count'])
df_exp[df_exp['name'].isin(dff[dff['count'] > 1]['name'])]
```
