```python
import pandas as pd
```

```python
fit = pd.read_csv('data/adv/fact_internet_sales.csv')
fit.head(3)
```

```python
product = pd.read_csv('data/adv/dim_product.csv')
product[product['ProductKey'] == 346]
```

```python
customer = pd.read_csv('data/adv/dim_customer.csv')
customer[customer['CustomerKey'] == 28389]
```

```python
currency = pd.read_csv('data/adv/dim_currency.csv')
currency.query('CurrencyKey == 39')
```

```python
fit_joined = fit.merge(product, on='ProductKey', how='inner')
columns = ['EnglishProductName', 'CustomerKey', 'CurrencyKey', 'OrderDateKey', 'UnitPrice']
fit_joined = fit_joined[columns]
```

```python
df1 = fit_joined.merge(currency, on='CurrencyKey').drop(
    columns=['CurrencyKey', 'CurrencyAlternateKey']
).merge(
    customer, on='CustomerKey'
)[
    ['EnglishProductName','OrderDateKey','UnitPrice','CurrencyName',	'FirstName', 'LastName']
]
```

```python
df1.groupby('CurrencyName')['UnitPrice'].agg(['mean'])
```
